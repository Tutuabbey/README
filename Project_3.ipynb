# **Model selection and training**

```python
#	Using Linear regression to estimate the relationship between the 2 variables chosen

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("bad-drivers.csv")
```

```python
# Choose our exploratory and target series + state name series

df = df[["State", "Car Insurance Premiums ($)", "Losses incurred by insurance companies for collisions per insured driver ($)"]]
df
```

Plotting the 2 variables against each other to visualize it

```python
df.plot(x= "Car Insurance Premiums ($)", y="Losses incurred by insurance companies for collisions per insured driver ($)", style="o")
```

Splitting into train and test samples

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df[["Car Insurance Premiums ($)", "Losses incurred by insurance companies for collisions per insured driver ($)"]], train_size=0.7, test_size=0.3)
print("Number of records in train sample:", len(train_df))
print("Number of records in test sample:", len(test_df))
```

```python
train_df.describe()
```

splitting the train and test dataframes into the exploratory (X) and the target (y) data sets.

```python
X_train = train_df["Car Insurance Premiums ($)"].values.reshape(-1, 1) #.values.reshape(-1, 1) is used when you only have one exploratory variable!
y_train = train_df["Losses incurred by insurance companies for collisions per insured driver ($)"]

X_test = test_df["Car Insurance Premiums ($)"].values.reshape(-1, 1)
y_test = test_df["Losses incurred by insurance companies for collisions per insured driver ($)"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

Measuring model performance using the Coefficient of Determination (R^2)

```python
print("R^2 score on training set:", model.score(X_train, y_train)) #df["Car Insurance Premiums ($)"].corr(df["Losses incurred by insurance companies for collisions per insured driver ($)"])
```

Measuring model performance using the Mean Absolute Error (MAE) metric

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_train_pred = model.predict(X_train)

print("MAE score on training set:", mean_absolute_error(y_train, y_train_pred))
```

plotting the training data on a graph and overlay the Linear Regression model

```python
ax = train_df.plot(x="Car Insurance Premiums ($)", y="Losses incurred by insurance companies for collisions per insured driver ($)", style="o") # Plot the data - elo98 vs elo15 from the dataframe

# Create the best fit line and plot it over the graph
# Create an array starting at the lowest elo98 score and finishing at the highest elo98 score. Create 1000 incremental steps between the two.
xx = np.linspace(train_df["Car Insurance Premiums ($)"].min(), train_df["Car Insurance Premiums ($)"].max(), 1000)

# We are going to use the equation y = b0 + b1*x to predict values of y (ELO15) using the 1000 values of x (ELO98) above
yy = model.intercept_ + model.coef_[0] * xx

# Plotting the line over the data
ax.scatter(xx, yy, color="red", marker='.')
```
